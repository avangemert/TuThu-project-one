```python
city_state = pd.DataFrame(school_shootings_df["Location"].str.split(',').tolist(),columns=["City","State"])
city_state = city_state.sort_values("State")
state_stats = pd.DataFrame(city_state["State"].value_counts())
state_stats = state_stats.reset_index()
state_stats = state_stats.sort_values("index")
state_stats["Politics"] = ["Red","Red","Red","Red","Blue","Blue","Blue","Split","Red","Blue","Blue","Split","Red","Red","Red",
                          "Blue","Blue","Blue","Blue","Red","Red","Red","Split","Blue","Blue","Blue","Red","Split","Red","Blue",
                          "Blue","Red","Red","Red","Blue","Blue","Blue","Red","Blue","Red"]
state_stats["Victims"] = school_zip_df["Bullet Hits"]
state_stats.head()
```

```python
red_states = state_stats.loc[state_stats["Politics"]=="Red"]
red_shootings = red_states["State"].sum()
blue_states = state_stats.loc[state_stats["Politics"]=="Blue"]
blue_shootings = blue_states["State"].sum()
split_states = state_stats.loc[state_stats["Politics"]=="Split"]
split_shootings  = split_states["State"].sum()
```

```python
shooting_politics = [red_shootings,blue_shootings,split_shootings]
labels = ["Republican","Democratic","Split"]
colors = ["red","lightblue","yellow"]
plt.pie(shooting_politics, labels=labels, colors=colors,autopct="%1.1f%%", shadow=True, startangle=40)
plt.title("Shootings by Voting Results (Last 4 Elections)")
plt.show()
```

```python
import pandas as pd
```

```python
# File to Load
shootings_path = "school_shootings_data.csv"

# Read the modified GoodReads csv and store into Pandas DataFrame
shootings_df = pd.read_csv(shootings_path, encoding="utf-8")
shootings_df.head()
```

```python
shootings_group_type = shootings_df.groupby(["School Type"])
shootings_group_type_mean = shootings_group_type.mean()
shootings_group_type_mean
```

```python
shootings_group_gender = shootings_df.groupby(["Shooter Gender"])
shootings_group_gender_mean = shootings_group_gender.mean()
shootings_group_gender_mean
```

```python
shootings_group_suicide = shootings_df.groupby(["Suicide"])
shootings_group_suicide_mean = shootings_group_suicide.mean()
shootings_group_suicide_mean
```

```python
shootings_group_former = shootings_df.groupby(["Former Student"])
shootings_group_former_mean = shootings_group_former.mean()
shootings_group_former_mean
```

```python
shootings_group_number = shootings_df.groupby(["Number of Attackers"])
shootings_group_number_mean = shootings_group_number.mean()
shootings_group_number_mean
```

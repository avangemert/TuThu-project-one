```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

census_data_df = pd.read_csv("Shootings_economics.csv")
school_zip_df = pd.read_csv("school_shootings_data_zip.csv")
school_shootings_df = pd.read_csv("school_shootings_data.csv")
school_shootings_df.head()
```

```python
school_shootings_df["Date"] = pd.to_datetime(school_shootings_df["Date"])
school_shootings_df
```

```python
# Pie chart of shooter gender

male_shooter_df = school_shootings_df.loc[school_shootings_df["Shooter Gender"]=="M"]
male_shooter_count = len(male_shooter_df["Shooter Gender"])
female_shooter_df = school_shootings_df.loc[school_shootings_df["Shooter Gender"]=="F"]
female_shooter_count = len(female_shooter_df["Shooter Gender"])
shooter_gender_count = [male_shooter_count,female_shooter_count]
labels = ["Male","Female"]
colors = ["lightblue","pink"]
plt.pie(shooter_gender_count, labels=labels, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=40)
plt.title("Shooter Gender Breakdown")
plt.show()
```

```python
# Number of Suicides after shooting pie chart

no_suicide_df = school_shootings_df.loc[school_shootings_df["Suicide"]=="N"]
yes_suicide_df = school_shootings_df.loc[school_shootings_df["Suicide"]=="Y"]
no_suicide = len(no_suicide_df["Suicide"])
yes_suicide = len(yes_suicide_df["Suicide"])
suicide = [no_suicide,yes_suicide]
labels = ["No","Yes"]
colors = ["lightblue","yellow"]
explode = [0.1,0]
plt.pie(suicide, explode=explode, labels=labels, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=40)
plt.title("Number of Suicides After Shooting")
plt.show()
```

```python
# Bar graph for types of school

elem_school_df = school_shootings_df.loc[school_shootings_df["School Type"]=="E"]
elem_school_count = len(elem_school_df["School Type"])
middle_school_df = school_shootings_df.loc[school_shootings_df["School Type"]=="M"]
middle_school_count = len(middle_school_df["School Type"])
high_school_df = school_shootings_df.loc[school_shootings_df["School Type"]=="H"]
high_school_count = len(high_school_df["School Type"])
college_df = school_shootings_df.loc[school_shootings_df["School Type"]=="U"]
college_count = len(college_df["School Type"])
school_type = [elem_school_count,middle_school_count,high_school_count,college_count]
labels = ["Elementary School","Middle School","High School","College"]
plt.bar(labels,school_type)
plt.title("Type of School")
plt.xlabel("Type of School")
plt.ylabel("Number of Shootings")
plt.show()
```

```python
labels = ["Elementary School","Middle School","High School","College"]
elem_school_shooter = elem_school_df["Shooter Age"].mean()
middle_school_shooter = middle_school_df["Shooter Age"].mean()
high_school_shooter = high_school_df["Shooter Age"].mean()
college_shooter = college_df["Shooter Age"].mean()
school_shooter_age = [elem_school_shooter,middle_school_shooter,high_school_shooter,college_shooter]
plt.bar(labels,school_shooter_age)
plt.title("Average Age of Shoooter per School Type")
plt.xlabel("Type of School")
plt.ylabel("Average Age of Shooter")
plt.show()
```

```python
# Pie chart of former students

yes_formstudent_df = school_shootings_df.loc[school_shootings_df["Former Student"]=="Y"]
yes_formstudent_count = len(yes_formstudent_df["Former Student"])
no_formstudent_df = school_shootings_df.loc[school_shootings_df["Former Student"]=="N"]
no_formstudent_count = len(no_formstudent_df["Former Student"])
former_student = [yes_formstudent_count,no_formstudent_count]
labels = ["Yes","No"]
colors = ["lightgreen","orange"]
explode = [0.1,0]
plt.pie(former_student, explode=explode, labels=labels, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=40)
plt.title("Former/Current Student Shooters")
plt.show()
```

```python
# Bar chart of age groups of shooters

bins = [0,9,19,29,39,49,99]
group_names = ["0-9","10-19","20-29","30-39","40-49","50+"]
shooter_age_groups = pd.cut(school_shootings_df["Shooter Age"], bins, labels=group_names)
school_shootings_df["Shooter Age Group"] = shooter_age_groups
pre_teen_shooter_df = school_shootings_df.loc[school_shootings_df["Shooter Age Group"] == "0-9"]
pre_teen_shooter_count = len(pre_teen_shooter_df["Shooter Age Group"])
teen_shooter_df = school_shootings_df.loc[school_shootings_df["Shooter Age Group"] == "10-19"]
teen_shooter_count = len(teen_shooter_df["Shooter Age Group"])
twenties_shooter_df = school_shootings_df.loc[school_shootings_df["Shooter Age Group"] == "20-29"]
twenties_shooter_count = len(twenties_shooter_df["Shooter Age Group"])
thirties_shooter_df = school_shootings_df.loc[school_shootings_df["Shooter Age Group"] == "30-39"]
thirties_shooter_count = len(thirties_shooter_df["Shooter Age Group"])
forties_shooter_df = school_shootings_df.loc[school_shootings_df["Shooter Age Group"] == "40-49"]
forties_shooter_count = len(forties_shooter_df["Shooter Age Group"])
old_shooter_df = school_shootings_df.loc[school_shootings_df["Shooter Age Group"] == "50-99"]
old_shooter_count = len(old_shooter_df["Shooter Age Group"])
age_groups = [pre_teen_shooter_count,teen_shooter_count,twenties_shooter_count,thirties_shooter_count,forties_shooter_count,old_shooter_count]
plt.bar(group_names,age_groups)
plt.title("Shooter Age Group")
plt.xlabel("Age Groups")
plt.ylabel("Number of Shooters")
plt.show()
```

```python
total_deaths = school_shootings_df["Deaths"].sum()
total_injuries = school_shootings_df["Injuries"].sum()
total_victims = total_deaths + total_injuries
victims = [total_deaths,total_injuries]
labels = ["Deaths","Injuries"]
colors = ['red','lightblue']
explode = [0.1,0]
plt.pie(victims, explode=explode, labels=labels, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=140)
plt.title("Percentage of Deaths to Injuries")
plt.show()
```

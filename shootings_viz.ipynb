```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

shootings_df = pd.read_csv("school_shootings.csv")
shootings_df
```

```python
mass_shootings = shootings_df.loc[shootings_df["Injuries"] + shootings_df["Deaths"] >= 3]
mass_shootings["Date"] = pd.to_datetime(mass_shootings["Date"])
mass_shootings
```

```python
num_shootings = len(shootings_df.index)
num_mass_shootings = len(mass_shootings.index)
num_mass_shootings
perc_mass_shootings = (num_mass_shootings / num_shootings) *100
perc_mass_shootings
```

```python
total_deaths = shootings_df["Deaths"].sum()
total_deaths
```

```python
total_injuries = shootings_df["Injuries"].sum()
total_injuries
```

```python
total_victims = total_deaths + total_injuries
total_victims
```

```python
victims = [total_deaths,total_injuries]
labels = ["Deaths","Injuries"]
colors = ['red','lightblue']
explode = [0.1,0]
plt.pie(victims, explode=explode, labels=labels, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=140)
plt.title("Percentage of Deaths to Injuries")
```
